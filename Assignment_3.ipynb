#Question 1
1. Basic Line Plot

   o Plot a simple line graph of ùë¶ = ùë•2 for ùë• values ranging from -10 to 10.

   o Label the x-axis as "X V

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 400)
y = x**2
plt.plot(x, y)
plt.xlabel('X Values')
plt.ylabel('Squared Values')

plt.title('Plot of y = x^2')

plt.show()
```

2. Bar Plot

'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].

#Question 3
3. Scatter Plot

   o Generate two lists of 20 random integers each (for x and y coordinates).

   o Create a scatter plot with these values, and customize it by adding color,
size, and labels to the points.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(0, 100, 20)
y = np.random.randint(0, 100, 20)
plt.scatter(x, y, c='blue', s=100, alpha=0.5, edgecolors='w', label='Data Points')

plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')

plt.title('Scatter Plot of Random Integers')
plt.legend()

plt.show()
```

#Question 4
4. Histogram

     o Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.

     o Customize the number of bins and add a grid to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)

plt.hist(data, bins=20, edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 100 Random Values from a Normal Distribution')

plt.grid(True)

plt.show()
```

#question 5
5. Pie Chart

    o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
with values [20, 30, 25, 15, 10].

    o Display the percentage values on the chart and use a color palette for
differentiation.

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Percentage Distribution of Categories')
plt.show()
```

#Question 6
6. Box Plot (Seaborn)

   o Create a box plot to visualize the distribution of the marks scored by
students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].
   
   o Highlight the median and any outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
sns.boxplot(data=marks)
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Distribution of Marks Scored by Students')
plt.show()
```

#Question 7
7. Heatmap (Seaborn)

    o Create a 5x5 matrix with random integers between 1 and 10 using
np.random.randint() and visualize it as a heatmap using Seaborn.
    
    o Add a color bar and annotate the heatmap with the values.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
matrix = np.random.randint(1, 11, size=(5, 5))

sns.heatmap(matrix, annot=True, cmap='viridis', cbar=True)
plt.title('Heatmap of Random Integers')
plt.show()
```

#Question 8
8. Comparison of Matplotlib and Seaborn

    o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.
    
    o Highlight the differences in the code, syntax, and aesthetics of the two
libraries.

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
plt.bar(categories, values)

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart using Matplotlib')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
import pandas as pd
data = pd.DataFrame({'Categories': categories, 'Values': values})
sns.barplot(x='Categories', y='Values', data=data)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart using Seaborn')
plt.show()
```

#Question 9
9. Pair Plot (Seaborn)

    o Manually create a small dataset with columns such as 'Age', 'Height',
'Weight' for 10 individuals and visualize the relationships using a pair plot.

```python
import seaborn as sns
import pandas as pd
data = pd.DataFrame({
    'Age': [23, 25, 28, 22, 30, 24, 27, 26, 29, 21],
    'Height': [165, 170, 175, 160, 180, 168, 172, 169, 177, 162],
    'Weight': [65, 70, 75, 60, 80, 68, 72, 69, 77, 63]
})

sns.pairplot(data)
plt.suptitle('Pair Plot of Age, Height, and Weight', y=1.02)

plt.show()
```

10.
Save and Style Plots

     ‚Ä¢ Plot any graph of your choice and save it in both .png and .pdf formats.
     
     ‚Ä¢ Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain
how the style impacts the visualization.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)

plt.xlabel('X Values')
plt.ylabel('Sine of X')
plt.title('Sine Wave')
plt.savefig('sine_wave.png')
plt.savefig('sine_wave.pdf')
plt.show()
```
